```python
# leo el df, y le asigno al index la columna 0
df=pd.read_csv("./pune_1965_to_2002.csv", index_col=0)
df.head()
```

```python
# a x , le asigno el valor de los indices, que en este caso son los años
x=df.index
x
```

```python
# creo una variable y, que le asigno la suma del df, pero por columnam en otras palabas
y=df.sum(axis="columns")
y
```

```python

# creo una figura con el tamaño 12/4
fig, ax =plt.subplots(figsize=(12,4))

# al ax, le paso los valores que cree antes (x e y) y que el label, las lineas se llamen ...
ax.plot(x,y, label="Precipitaciones Acumuladas")

# titulos al ejex 
ax.set_xlabel("Año")

# titulos al ejey
ax.set_ylabel("Precipitaciones en MM")

# titulo al grafico
ax.set_title("Precipitaciones Acumuladas por año")

# muestra las leyendas dentro del grafico.
ax.legend()
```

```python
# igual que el grafico anterior, nada mas que en vez de lineas, muestro circulos.

# tamanli de la figura
fig, ax = plt.subplots(figsize=(12, 4))
# paso valores
ax.scatter(x, y, label='Precipitaciones acumuladas')

#titulos a los ejes
ax.set_xlabel('Año')
ax.set_ylabel('Precipitación acumulada (mm.)')

# titulos al grafico
ax.set_title('Precipitaciones acumuladas según el año')

# muestro leyenda
ax.legend()
```

```python
# creo figuras
fig, ax = plt.subplots()
x_agos=df['Aug']
y_sep=df['Sep']

#creo una variable, con graficos de dispersion, circulos, 
# el parametro c toma secuencia de colores, en este caso años.
mapeo=ax.scatter(x_agos,y_sep, c=df.index)

# Esta función crea una barra de color asociada a una figura. Toma como argumento el 
# objeto de dispersión (mapeo en este caso) que se desea asociar a la barra de color.
fig.colorbar(mapeo)
ax.set_title("Precipitaciones en Agosto y Septiembre")
ax.set_xlabel("Precipitaciones en Agosto")
ax.set_ylabel("Precipitaciones en Septiemnbre")
```

```python
# creo el grafico
fig, ax=plt.subplots(figsize=(12,3))

# creo una linea, con parametro año, y la lluvia de enero, paso leyenda, y color verde
ax.plot(df.index, df['Jan'], label="Precipitaciones de Enero", color="green")
# creo una linea, con parametro año, y la lluvia de febrero, paso leyenda, y color verde
ax.plot(df.index, df['Feb'], label="Precipitaciones de Febrero", color="red")

# creo dos variables con los maz de enero y febrero
max_enero=df['Jan'].max()
max_febrero=df['Feb'].max()

# creo dos lineas horizontales, pasando el valor maximo de enero y febrero, paso tipo de linea, opacidad, tamaño, y leyenda
ax.axhline(max_enero,   color="blue",  linestyle="--", alpha=0.5, linewidth=3, label="Maxima de Enero")
ax.axhline(max_febrero, color="black",linestyle=":", alpha=0.5, linewidth=3, label="Maxima de Febrero")

# titulos a los graficos
ax.set_xlabel("Año")
ax.set_ylabel("Precipitaciones en MM")
ax.set_title("Precipitaciones en enero y febrero.")

# muestro todos los años- con esta formula ( osea que los limites vayan de 0 al ultimo.)
ax.set_xlim(df.index[0], df.index[-1])
ax.legend()




```

 Piechart para ver distribucion de empleados por departamento

```python
# cuento cuantos empleados hay por departamentos.
df2_departamentos=df2[['EmpID','Department']].groupby(by='Department').count()
df2_departamentos
```

```python

# creo figura
fig, ax=plt.subplots(figsize=(10,6))

# creo un grafico de torta, le paso los parametros,  y que muestre por el indice(tipos de trabajos)
# el autopct es como muestra los numero.s
ax.pie(df2_departamentos.EmpID, labels=df2_departamentos.index, autopct="%1.1f%%")

# titulos.
ax.set_title("Distribucion por departamentos.")
```
