```python
# creo un histplot, le paso el df
# al eje x le paso el tamaÃ±o del pico, muestro speciies (hue), bins 12
# multiple por stack (las barras se apilen una arriba de la otra.)
sns.histplot(data=pinguino, x="flipper_length_mm", hue="species", multiple="stack", bins=12)
```

```python
# crea figura con ejes (graficos)(axs)
# gridspec --> usa un dict, para darle dif anchos a las columnas.
f, axs=plt.subplots(1,2, figsize=(8,4), gridspec_kw=dict(width_ratios=[4,3]))


# creo grafico de dispersion, paso ejex e y , con hue , y ocupa el primer eje
sns.scatterplot(data=pinguino, x="flipper_length_mm", y="bill_length_mm", hue="species", ax=axs[0])

# creo un histograma, paso parametors, shrink (separacion de barras), alpha (opacidad) sin leyenda,
# y que vaya a la posicion dos
sns.histplot(data=pinguino, x="species",hue="species",shrink=0.8, alpha=0.8, legend=False, ax=axs[1])

# ajusta diseÃ±os. para evitar superposiciones.
f.tight_layout()
```

```python
vuelos=sns.load_dataset("flights")
vuelos.head()
```

```python
# creo un df, con la funcion pivot
# 1er parametro va como fila
# 2do como columna
# 3ero como valor
vuelos_wide=vuelos.pivot("year","month", "passengers")
vuelos_wide
sns.lineplot(data=vuelos_wide)
```

```python
df_ejercicio = sns.load_dataset('exercise')
df_ejercicio = df_ejercicio.drop('Unnamed: 0', axis='columns')
df_ejercicio.head()
```

```python
df_30_min = df_ejercicio[df_ejercicio['time'] == '30 min']
df_30_min.head()
```

```python
# Figure -level
ax = sns.displot(data=df_30_min, x='pulse', kind='kde', hue='kind', fill=True)
ax.set(xlabel='Frecuencia CardÃ­aca', ylabel='Densidad', title='DistribuciÃ³n  de las pulsaciones')
plt.show()
```

```python
ax = sns.catplot(data=df_ejercicio, kind='violin', x='time', y='pulse', hue='diet', split=True)
ax.set(xlabel='DuraciÃ³n de ejercicio', ylabel='Frecuencia cardÃ­aca', title='CategorizaciÃ³n de la distribuciÃ³n de pulsaciones')
plt.show()
```
