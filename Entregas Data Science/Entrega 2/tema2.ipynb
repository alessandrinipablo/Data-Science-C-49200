Este llamado Clima Global proporciona información detallada sobre condiciones climáticas en diversas ubicaciones geográficas, incluyendo datos de temperatura, viento, presión, precipitación, humedad, visibilidad y calidad del aire. También incluye detalles de ubicación, zona horaria, y fases lunares. 

Explicacion de lo que significa cad columnas (variable)

country: El país al que pertenecen los datos climáticos.
location_name: El nombre de la ubicación o lugar específico para el que se proporcionan los datos climáticos.
latitude: La latitud geográfica de la ubicación.
longitude: La longitud geográfica de la ubicación.
timezone: La zona horaria de la ubicación.
last_updated_epoch: La marca de tiempo en formato epoch (número de segundos desde el 1 de enero de 1970) de la última actualización de los datos.
last_updated: La fecha y hora de la última actualización de los datos.
temperature_celsius: La temperatura en grados Celsius.
temperature_fahrenheit: La temperatura en grados Fahrenheit.
condition_text: La descripción textual de las condiciones climáticas (por ejemplo, "soleado", "lluvioso").
wind_mph: La velocidad del viento en millas por hora.
wind_kph: La velocidad del viento en kilómetros por hora.
wind_degree: La dirección del viento en grados.
wind_direction: La dirección del viento en texto (por ejemplo, "norte", "sur").
pressure_mb: La presión atmosférica en milibares.
pressure_in: La presión atmosférica en pulgadas de mercurio.
precip_mm: La cantidad de precipitación en milímetros.
precip_in: La cantidad de precipitación en pulgadas.
humidity: El porcentaje de humedad relativa del aire.
cloud: La cobertura de nubes en porcentaje.
feels_like_celsius: La sensación térmica en grados Celsius.
feels_like_fahrenheit: La sensación térmica en grados Fahrenheit.
visibility_km: La visibilidad en kilómetros.
visibility_miles: La visibilidad en millas.
uv_index: El índice UV (Ultravioleta) que indica la intensidad de la radiación ultravioleta.
gust_mph: Las ráfagas de viento en millas por hora.
gust_kph: Las ráfagas de viento en kilómetros por hora.
air_quality_Carbon_Monoxide: La calidad del aire en términos de concentración de monóxido de carbono.
air_quality_Ozone: La calidad del aire en términos de concentración de ozono.
air_quality_Nitrogen_dioxide: La calidad del aire en términos de concentración de dióxido de nitrógeno.
air_quality_Sulphur_dioxide: La calidad del aire en términos de concentración de dióxido de azufre.
air_quality_PM2.5: La calidad del aire en términos de concentración de partículas PM2.5.
air_quality_PM10: La calidad del aire en términos de concentración de partículas PM10.
air_quality_us-epa-index: El índice de calidad del aire según la normativa de la Agencia de Protección Ambiental de EE. UU.
air_quality_gb-defra-index: El índice de calidad del aire según la normativa del Departamento de Medio Ambiente, Alimentación y Asuntos Rurales del Reino Unido.
sunrise: La hora de salida del sol.
sunset: La hora de puesta del sol.
moonrise: La hora de la salida de la luna.
moonset: La hora de la puesta de la luna.
moon_phase: La fase lunar actual (por ejemplo, "creciente", "menguante").
moon_illumination: La iluminación de la luna en porcentaje.
