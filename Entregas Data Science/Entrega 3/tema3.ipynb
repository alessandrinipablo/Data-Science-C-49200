```python
import pandas as pd 
import numpy as np
```

El presente DataSet, es sobre las reservas de propiedad mediante la App Airbnb en la ciudad de Nueva York.

id: El identificador único del anuncio de Airbnb.
NAME: El nombre del alojamiento o propiedad.
host id: El identificador único del dueño o encargado del alojamiento.
host_identity_verified: Indica si la identidad del dueño o encargado  ha sido verificada (puede ser "Verificado" o "No verificado").
host name: El nombre del dueño o encargado .
neighbourhood group: El grupo de vecindario (barrio) al que pertenece la propiedad.
neighbourhood: El vecindario o barrio donde se encuentra la propiedad.
lat: La latitud geográfica de la ubicación.
long: La longitud geográfica de la ubicación.
country: El país donde se encuentra la propiedad.
country code: El código de país de la propiedad.
instant_bookable: Indica si la propiedad está disponible para reserva instantánea (puede ser "Sí" o "No").
cancellation_policy: La política de cancelación para la reserva.
room type: El tipo de habitación o alojamiento.
Construction year: El año de construcción de la propiedad.
price: El precio de alquiler por noche.
service fee: La tarifa de servicio adicional.
minimum nights: El número mínimo de noches requeridas para la reserva.
number of reviews: El número total de reseñas o comentarios de los huéspedes.
last review: La fecha de la última reseña.
reviews per month: El promedio de reseñas por mes.
review rate number: La calificación promedio basada en las reseñas de los huéspedes.
calculated host listings count: El número de propiedades listadas por el mismo dueño o encargado .
availability 365: El número de días disponibles para reservar en un año.
house_rules: Las reglas de la casa establecidas por el anfitrión.
license: Licencia o permiso legal para operar la propiedad como alquiler de Airbnb.
