```python
import pandas as pd 
import numpy as np 
```

Este data set, que trata sobre el futbol argentino, muestra datos de los ultimos 7 años, mostrando temporadas, estadios, resultados, goles locales y goles visitantes. Tambien muestra un monton de informacion , como por ejemplo posicion adelantada, tarjetas amarillas y rojas, elementos muy importantes para analizar, graficar y utilizar para predecir resutaldos.


 Variables 

- temporada: Temporada del fútbol.
- semana: Semana en que se jugó el partido.
- dia: Día de la semana del partido.
- fecha: Fecha del partido.
- hora: Hora de inicio del partido.
- local: Equipo local.
- visitante: Equipo visitante.
- sedes: Lugar del partido.
- arbitro: Nombre del árbitro.
- resultado: Resultado del partido.
- local_posesion: Posesión del balón del equipo local.
- visitante_posesion: Posesión del balón del equipo visitante.
- local_goles: Goles del equipo local.
- visitante_goles: Goles del equipo visitante.
- local_asistencias: Asistencias del equipo local.
- local_goles_penal: Goles de penal del equipo local.
- local_penales: Penales ejecutados por el equipo local.
- local_disparos: Disparos del equipo local.
- local_disparos_puerta: Disparos a puerta del equipo local.
- local_amarillas: Tarjetas amarillas al equipo local.
- local_rojas: Tarjetas rojas al equipo local.
- local_faltas_cometidas: Faltas cometidas por el equipo local.
- local_faltas_recibidas: Faltas recibidas por el equipo local.
- local_posiciones_adelantadas: Posiciones adelantadas del equipo local.
- local_pases_cruzados: Pases cruzados del equipo local.
- local_derribos: Derribos del equipo local.
- local_intercepciones: Intercepciones del equipo local.
- local_goles_contra: Goles en contra del equipo local.
- visitante_asistencias: Asistencias del equipo visitante.
- visitante_goles_penal: Goles de penal del equipo visitante.
- visitante_penales: Penales ejecutados por el equipo visitante.
- visitante_disparos: Disparos del equipo visitante.
- visitante_disparos_puerta: Disparos a puerta del equipo visitante.
- visitante_amarillas: Tarjetas amarillas al equipo visitante.
- visitante_rojas: Tarjetas rojas al equipo visitante.
- visitante_faltas_cometidas: Faltas cometidas por el equipo visitante.
- visitante_faltas_recibidas: Faltas recibidas por el equipo visitante.
- visitante_posiciones_adelantadas: Posiciones adelantadas del equipo visitante.
- visitante_pases_cruzados: Pases cruzados del equipo visitante.
- visitante_derribos: Derribos del equipo visitante.
- visitante_intercepciones: Intercepciones del equipo visitante.
- visitante_goles_contra: Goles en contra del equipo visitante.
- local_puntos_acumulados_temporada: Puntos acumulados por el equipo local en la temporada.
- visitante_puntos_acumulados_temporada: Puntos acumulados por el equipo visitante en la temporada.
