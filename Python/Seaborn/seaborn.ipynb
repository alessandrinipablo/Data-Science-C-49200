```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

```

tipos de graficas de seaborn
catplot --> categoricas. ( cuando analizo una categoria, tipo de fruta y su venta)
displot --> distribuciones. (ejemplo datos de altura de estudiantes)
relplot --> relacional.( relacion entre variables, horas de estudio con notas de los estudiantes.)

categoricas se subdivide mas 
-dispersion:

1-strip-->stripplot(Representa puntos individuales en un solo eje, ideal para conjuntos de datos pequeños.)
2-swarm-->swarmplot(Similar a stripplot, pero los puntos se ajustan para evitar la superposición)

-distribucion
1-box-->boxplot(Muestra la distribución de datos en términos de cuartiles.)
2-boxen-->boxenplot(Similar al boxplot, pero más detallado y adecuado para conjuntos de datos más grandes)
3-violin-->violinplot(Combina la información de un boxplot con una estimación de la densidad de kernel  en     cada lado)

-Estimatacion.
1-point-->pointplot(Muestra la relación central entre dos variables categóricas.)
2-var-->varplot(Muestra la variabilidad en la relación entre dos variables categóricas)
3-count-->countplot(Muestra la cuenta de observaciones en cada categoría.)

GRAFICAS DE DISPERSION

```python
#configuro seaborn
# configuro estilo en ticks
# --> otros tipos ("darkgrid", "whitegrid", "dark", "white", "ticks")
# context talk --> apto para presentaciones.
# palette --> colores de los graficos.
sns.set_theme(style="ticks", context="talk", palette="tab10")

# levanto el dataframe ya cargado, tips, que son propinas.
consumos=sns.load_dataset("tips")
# renombro columnas.
consumos.columns=["total","propina","genero","fumador","dia","tipo","comensales"]
consumos
```

GRAFICA STRIPPLOT

```python
# sns. catplot --> metodo que permite crear todas las graficas categoricas.

# paso los parametros de eje x e y , y de donde lo va a sacar

# kind--> strip (kind es forma de grafico)

# hue --> matiz (evaluo por una variable, en este caso comensales )

# s--> tamaño de los puntos.
sns.catplot(x="dia",y="total", data=consumos, kind="strip", hue="comensales", col="fumador", s=10)

# poniendole col ( que es fumador), analiza en base a dos , fumador y no fumador.

# jitter lo que hace es poner los datos en linea recta con jitter false
# para que no se mezclen los puntos, sino que esten separados entre si

# muestro
plt.show()
```

GRAFICA SWARMPLOT (COLMENA DE ABEJAS.)

```python
# CAMBIO EL TIPO DE KIND por swarm, lo que hace este, evita superposicion.
# se utliza cuando tenemos pocos datos.
sns.catplot(x="dia",y="total", data=consumos, kind="swarm", hue="comensales", col="fumador")
plt.show()
```

GRAFICAS CATEGORICAS  DE DISTRIBUCION.

GRAFICA BOXPLOT

```python


# mismo codigo que venimos haciendo, cambio  el kind.
# Analizando los graficos
# --> las puntas representan los valores minimos y maximos.
# --> dentro de la caja la linea recta horizontal representa la mediana (valor que se encuentra en la mitad).
# --> las lineas representan los cuartiles, linea 1 cuartil 1, la mediana cuartil 2, la tercer linea cuartil 3
# los puntos indican --> datos outliers, o atipico (fuera de lo normal.)
sns.catplot(x="dia",y="total",data=consumos,kind="box",col="fumador")
plt.show()
```

GRAFICA BOXENPLOT

```python

# grafica igual que la anterior, nada mas que se usa cuando hay muchos datos.
sns.catplot(x="dia",y="total",data=consumos,kind="boxen",col="fumador")
plt.show()
```

GRAFICA VIOLINPLOT

```python
# muy parecido al grafico BOXPLOT, REPRESENTA MEDIANAS, MIN MAX, Y LOS CUARTILES, NADA MAS QUE CON FORMA 
# DE VIOLIN.

# tambien puedo agregar que me lo defina en fumador o no fumador.

# el split, lo que hace es juntarlo o no 
sns.catplot(x="dia",y="total",data=consumos,kind="violin")
sns.catplot(x="dia",y="total",data=consumos,kind="violin",hue="fumador")
sns.catplot(x="dia",y="total",data=consumos,kind="violin",hue="fumador", split=True)
plt.show()
```

Graficas Categoricas de Estimacion.


Grafica Point Plot

```python

# los puntos donde unen, son los promedios de los consumos.}
# tambien se divide por fumadores o no fumadores. 
# las lineas represetan el intervalo de confianza, osea donde ocurre el 95% de los hechos.
sns.catplot(x="dia",y="total", data=consumos, kind="point", col="fumador")
plt.show()
```

Grafica BarPlot

```python
# analizo las graficas por comensales.
sns.catplot(x="dia",y="total", data=consumos, kind="bar", hue="comensales")
plt.show()
```

Grafico CountPlot (grafica de conteo)

```python
sns.catplot(x="dia", data=consumos, kind="count", hue="comensales")
plt.show()
```
